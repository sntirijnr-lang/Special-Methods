### 1. Constructor (`__init__`)
This is Python’s standard constructor. Whenever you create a new instance of a class, this runs automatically to set up its initial attributes. You don't call it manually; passing arguments to the class name triggers it.

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

my_book = Book("Things Fall Apart", "Chinua Achebe")
print(my_book.title)
```

### 2. String Representation (`__str__` and `__repr__`)
Both of these control what gets printed when someone turns your object into a string, but they target different audiences:
* __str__ is for regular users (clean, readable output). Called by print(obj).
* __repr__ is for developers (unambiguous, official representation). Called when you just type the object name in the console. If __str__ is missing, Python falls back to __repr__.

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f"Point at ({self.x}, {self.y})"

    def __repr__(self):
        return f"Point(x={self.x}, y={self.y})"

p = Point(4, 5)
print(str(p))   # Triggers str
print(repr(p))  # Triggers repr
```

### 3. Length (`__len__`)
This lets custom objects respond to the built-in len() function. Without it, calling len(my_object) throws a TypeError. It must always return a non-negative integer.

```python
class Playlist:
    def __init__(self, songs):
        self.songs = songs

    def __len__(self):
        return len(self.songs)

roadtrip_jams = Playlist(["Blinding Lights", "Last Last", "Unavailable"])
print(len(roadtrip_jams))
```

### 4. Equality (`__eq__`)
By default, the == operator only checks if two variables point to the exact same object in memory. Defining __eq__ lets us change that behavior so == actually compares the *data* inside two different objects.

```python
class Coin:
    def __init__(self, value):
        self.value = value

    def __eq__(self, other):
        if isinstance(other, Coin):
            return self.value == other.value
        return False

coinA = Coin(50)
coinB = Coin(50)

print(coinA == coinB) # Returns True because of our eq logic
```

### 5. Operator Overloading (`__add__` and `__sub__`)
These allow custom objects to interact with standard math symbols. __add__ defines what happens when you use the + operator on your object, and __sub__ defines the - operator.

```python
class Wallet:
    def __init__(self, balance):
        self.balance = balance

    def __add__(self, amount):
        return Wallet(self.balance + amount)

    def __sub__(self, amount):
        return Wallet(self.balance - amount)

my_wallet = Wallet(100)
payday = my_wallet + 50
tax = payday - 20

print(payday.balance)
print(tax.balance)
```

### 6. Item Access (`__getitem__`)
This gives your object indexing capabilities, making it act like a list or a dictionary. It allows you to grab data out of the object using square bracket syntax: my_object[index].

```python
class Train:
    def __init__(self, carriages):
        self.carriages = carriages

    def __getitem__(self, index):
        return self.carriages[index]

express = Train(["Engine", "First Class", "Economy", "Caboose"])
print(express[1]) # Grabs index 1 via getitem
```

### 7. Item Assignment (`__setitem__`)
The companion to __getitem__. This lets you *assign* or overwrite a value inside your object using bracket syntax: my_object[key] = new_value.

```python
class Locker:
    def __init__(self):
        self.contents = {}

    def __setitem__(self, locker_number, item):
        self.contents[locker_number] = item

gym = Locker()
gym[104] = "Sneakers"
print(gym.contents)
```

### 8. Membership Testing (`__contains__`)
This powers Python's in keyword. When you write x in my_object, Python secretly runs my_object.__contains__(x) behind the scenes. It should return True or False.

```python
class Fridge:
    def __init__(self, items):
        self.items = items

    def __contains__(self, food):
        return food.lower() in [i.lower() for i in self.items]

my_fridge = Fridge(["Milk", "Eggs", "Leftover Pizza"])

print("Eggs" in my_fridge)
print("Apples" in my_fridge)
```

### 9. Callable Objects (`__call__`)
Normally you put parentheses () after a function. Defining __call__ allows an *instance of a class* to act like a function itself, meaning you can execute my_object().

```python
class Multiplier:
    def __init__(self, factor):
        self.factor = factor

    def __call__(self, number):
        return number * self.factor

double = Multiplier(2)

print(double(10)) # Calling the instance directly like a function
```

### 10. Iteration Support (`__iter__` and `__next__`)
These two work together to let your object loop inside a for loop:
* __iter__ returns the iterator object itself (usually `self`).
* __next__ fetches the next item. When the data runs out, it must raise a StopIteration exception to tell the loop to stop.

```python
class Countdown:
    def __init__(self, start):
        self.current = start

    def __iter__(self):
        return self

    def __next__(self):
        if self.current <= 0:
            raise StopIteration

        num = self.current
        self.current -= 1
        return num

for second in Countdown(3):
    print(second)
```

### 11. Boolean Evaluation (`__bool__`)
Defines the "truthiness" of an object when evaluated in an if statement. If it isn't defined, Python checks __len__() (0 is False, anything else is True). If neither exists, custom objects evaluate to True by default.

```python
class Battery:
    def __init__(self, percentage):
        self.percentage = percentage

    def __bool__(self):
        return self.percentage > 0

dead_phone = Battery(0)
alive_phone = Battery(45)

print(bool(dead_phone))
print(bool(alive_phone))
```

### 12. Destructor (`__del__`)
The opposite of __init__. This runs right before an object gets permanently destroyed by Python's garbage collector. It’s mostly used for cleanup tasks, like closing a file or shutting down a network connection.

```python
class TempFile:
    def __init__(self, filename):
        self.filename = filename
        print(f"Opened {self.filename}")

    def __del__(self):
        print(f"Cleanup: {self.filename} wiped from memory.")

doc = TempFile("data.csv")
del doc # Forcing the delete triggers del
```
