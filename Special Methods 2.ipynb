```python
##Simple Inheritance
class Person:
    def greet(self):
        print("Good day everyone!")

# Student inherits from Person using parenthetical syntax
class Student(Person):
    pass

# Testing the simple inheritance
s = Student()
s.greet()  # Output: Good day everyone!
```

```python
#Inheriting Attributes
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    pass

# Testing attribute inheritance
s = Student("Stephen")
print(s.name)  # Output: Matthew
```

```python
#Extending a Parent Class
class Person:
    def greet(self):
        print("Hello")

class Student(Person):
    # Extending the class with a new unique method
    def study(self):
        print("I’m Enjoying the Python OOP Course")

# Testing extension
daniella = Student()
daniella.greet()  # Inherited method
daniella.study()  # Unique child method
```

```python
#Constructor Inheritance
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, name, student_id):

        Person.__init__(self, name)
        self.student_id = student_id

student = Student("Stephen", "S12345")
print(student.name)
print(student.student_id)
```

```python
#The super() built-in function
class Person:
    def __init__(self, name):
        self.name = name

    def display(self):
        print("Name:", self.name)

class Student(Person):
    def __init__(self, name, student_id):
        super().__init__(name)
        self.student_id = student_id

    def info(self):
        print("ID:", self.student_id)

# Execute display and info
student = Student("Stephen", 1001)
student.display()
student.info()
```

```python
#Contructor Inheritance b
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, name, student_id):
        # Explicitly call the parent class constructor
        Person.__init__(self, name)
        self.student_id = student_id

# Testing explicit parent constructor call
student = Student("Stephen", "S12345")
print(student.name)       # Output: Matthew
print(student.student_id)  # Output: S12345
```

```python
#Method Overrinding
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    # This overrides the parent's speak method
    def speak(self):
        print("Dog barks")

# Testing overriding
dog = Dog()
dog.speak()  # Output: Dog barks
```

```python
#Calling the Parent Method
class Animal:
    def speak(self):
        print("Animal sound")

class Dog(Animal):
    def speak(self):
        super().speak()  # Runs parent's functionality first
        print("Dog barks")   # Adds specific child behavior

# Testing combined method execution
dog = Dog()
dog.speak()
```

```python
# Parent Class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")


# Child Class 1: Student
class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id

    def enroll_course(self, course_name):
        print(f"{self.name} has enrolled in {course_name}.")

    def display_info(self):
        super().display_info()
        print(f"Student ID: {self.student_id}")


# Child Class 2: Lecturer
class Lecturer(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)
        self.employee_id = employee_id

    def teach_course(self, course_name):
        print(f"{self.name} is teaching {course_name}.")


# Demonstration
if __name__ == "__main__":
    # Create one object of each child class
    student = Student("Stephen", 19, "FOE.41.006.128.25")
    lecturer = Lecturer("Dr. Matthew Cobbinah", 40, "E98765")

    print("=== Student Information (Overridden display_info) ===")
    student.display_info()
    student.enroll_course("Object Oriented Programming")

    print("\n=== Lecturer Information (Inherited display_info) ===")
    lecturer.display_info()
    lecturer.teach_course("Object Oriented Programming")
```
